```python
import numpy as np
import pandas as pd
```

```python
Laliga_DF=pd.read_csv("Laliga.csv",skiprows=[0])
Laliga_DF
```

```python
Laliga_DF.replace('-','0',inplace=True) # Replacing 0 with - dash
Laliga_DF
```

```python
# List down the Team names between 1930 to 1980
# Convert the column value as String and check if string starts with 1930 till 1980
Laliga_DF["Debut"]=Laliga_DF["Debut"].astype(str) 
lst_Search= Laliga_DF[Laliga_DF["Debut"].str.startswith('193')|Laliga_DF["Debut"].str.startswith('194')|Laliga_DF["Debut"].str.startswith('195')|Laliga_DF["Debut"].str.startswith('196')|Laliga_DF["Debut"].str.startswith('197')]['Team']
print("The Search results are" + "\t" + str(len(lst_Search)) + "\n")
print("\t" + "\t" + "Team"+ "\n")
print(lst_Search)
```

```python
#Convert the column type to Numeric and then sort the column Points
Laliga_DF['Points'] = Laliga_DF['Points'].apply(pd.to_numeric)
# descending logic apply
print(Laliga_DF[["Team","Points"]].sort_values("Points",ascending=False).nlargest(5,"Points"))
```

```python
def Goal_diff_count(_for,_Against,_TeamName):
    Goal_diff_Count_Var=_for-_Against # computes the difference of all teams
    Goal_diff_List=[Goal_diff_Count_Var,_TeamName] # List with goal difference and Teamname
    print ("Goal Difference" + "\t" + "Team Name")
    for i in range(len(Goal_diff_List[0])):  # prints goal difference and Teamname
        print(str(Goal_diff_List[0][i]) + "\t" + "\t"+ str(Goal_diff_List[1][i]))
    aa=list(Goal_diff_List[0])

    print("The Max Goal difference and the TEAM who scored is : ",  Goal_diff_List[0].max(), "   ==>  ", Goal_diff_List[1][aa.index(Goal_diff_List[0].max())])
    print("The Min Goal difference and the TEAM who scored is : ",  Goal_diff_List[0].min(), "   ==>  ", Goal_diff_List[1][aa.index(Goal_diff_List[0].min())])

    #return Goal_diff_List
```

```python
Laliga_DF['GoalsFor'] = Laliga_DF['GoalsFor'].apply(pd.to_numeric)
Laliga_DF['GoalsAgainst'] = Laliga_DF['GoalsAgainst'].apply(pd.to_numeric)
Goal_diff_count(Laliga_DF["GoalsFor"],Laliga_DF["GoalsAgainst"],Laliga_DF["Team"])
```

```python
try:
    Laliga_DF["GamesPlayed"]= Laliga_DF["GamesPlayed"].apply(pd.to_numeric)
    Laliga_DF["GamesWon"]= Laliga_DF["GamesWon"].apply(pd.to_numeric)
    Laliga_DF.insert(Laliga_DF.shape[1],"Winning Percent",(((Laliga_DF["GamesWon"])/(Laliga_DF["GamesPlayed"]))*100).round(decimals=2))
except ZeroDivisionError:
    Laliga_DF["Winning Percent"].fillna(0,inplace=True)
finally:
    Laliga_DF["Winning Percent"] = Laliga_DF["Winning Percent"].apply(pd.to_numeric)
    print("The top 5 team with highest winning percent are" + "\n")
    print(Laliga_DF[["Team","Winning Percent"]].sort_values("Winning Percent",ascending=False).nlargest(5,"Winning Percent"))
```
